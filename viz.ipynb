```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
sns.set(style="white",color_codes=True)
plt.rcParams['figure.figsize'] = (15,9.27)
matplotlib.rcParams['mathtext.fontset'] = "cm"
```

```python
def smooth_plot(x,y,area=False,number=300,color='b',ls='solid',lw=1):
    from scipy.interpolate import spline
    x_np,y_np = np.array(x),np.array(y)
    xnew = np.linspace(x_np.min(),x_np.max(),number)
    y_smooth = spline(x_np,y,xnew)
    if area == True:
        plt.fill_between(xnew,y_smooth,color=color)
    else:
        plt.plot(xnew,y_smooth,color=color,linestyle=ls,linewidth=lw)
```

```python
def latex(s=''):
    n = len(s.split(' '))
    if n == 1:
        if s == '':
            return '$'+'\ '+'$'
        else:
            return '$'+s+'$'
    else:
        return '$'+'\ '.join(s.split(' '))+'$'  
```

```python
np.random.seed(123)
x = np.arange(1,21)
y = np.random.randint(1,10,20)
smooth_plot(x,y,color='#81D8D0',ls='--',lw=3)
plt.xlim(1,20)
plt.ylim(0,10)
plt.xlabel(latex('Independent Varaible'),fontsize=15)
plt.ylabel(latex('dependent Varaible'),fontsize=15)
plt.title(latex('The Relationship Between This Two Variable'),fontsize=15)
```
