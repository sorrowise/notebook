```python
def missing(df):
    res = pd.DataFrame()
    res['num'] = df.isnull().sum()
    res['pert'] = 100*res['num']/sum(res['num'])
    res.sort_values('pert',ascending=False,inplace=True)
    res = res[res.num>0]
    return res
```
