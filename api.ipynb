```python
def fplot(func,beg,end,lw=2,xlab='x',ylab='y',title='',fs=20):
    '''
    Given form of the function, this code snippest can draw its image on a particular interval

    Args:
        func(function): the function need to be plotted
        beg(float): The left end of the function interval
        end(float): The right end of the function interval
        lw(float,optional,default=2): width of the line
        xlab(string,optional,default='x'): The title of the X axis
        ylab(string,optional,default='y): The title of the Y axis
        title(string,optional,default=''): The title of the plot
        fs(int,optional,default=20): fontsize of X axis, Y axis and title
    
    Returns:
        The function image on a particular interval
    '''
    x = np.arange(beg,end+0.001,0.001)
    y = np.array(list(map(func,x)))
    plt.plot(x,y,linewidth=lw)
    plt.xlim(beg,end)
    plt.xlabel(latex(xlab),fontsize=fs)
    plt.ylabel(latex(ylab),fontsize=fs)
    plt.title(latex(title),fontsize=fs)
```
