```python
np.random.seed(123)
x = np.random.uniform(-1,1,500)
y = np.random.uniform(-1,1,500)
z = x**2 + y**2 + (np.random.uniform(-1,1,500)/2)
data = pd.DataFrame({'x':x,'y':y,'z':z})
```

```python
def func(x):
    if x>1:
        return 0
    else:
        return 1
```

```python
data['target'] = data.z.apply(func)
```

```python
sns.scatterplot(x='x',y='y',style='target',hue='target',data=data)
xxx = np.linspace(-1,1,1000)
yyy = np.sqrt(1-xxx**2)
plt.plot(xxx,yyy,'g--')
plt.plot(xxx,-yyy,'g--')
```
